```python
import numpy as np
import os
from PIL import Image
from matplotlib import pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img,img_to_array
```

### Set Image generator

```python
ImageGen=ImageDataGenerator(
    rotation_range=48,
    width_shift_range=0.2,
    rescale=1./255,
    zoom_range=0.3,
    horizontal_flip=True,
    vertical_flip=True,
    shear_range=0.2,
    fill_mode='nearest'
)
```

### Load Image


```python
img_path="imggen/original/Thanu.jpg"
out_path="imggen/out/"
```

```python
pic=load_img(img_path);
pic.getpixel
```

### Image to Array

```python
pic_array=img_to_array(pic)
new_pic_array=pic_array.reshape((1,) + pic_array.shape)
print(pic_array.shape)
print(new_pic_array.shape)
```

```python
count=0
for batch in ImageGen.flow(new_pic_array,batch_size=1,save_to_dir=out_path,save_prefix='ThanuMahee',save_format='png'):
    count+=1
    if count==15:
        break
print("Geration Complete")
```

### Show all picture 

```python
gernators=os.listdir(out_path)
length=len(gernators)
```

```python
fig,axs=plt.subplots(nrows=3,ncols=6)
axs=axs.flatten()
for i in range(1,length):
    filepath=os.path.join(out_path,gernators[i])
    IMG=Image.open(filepath)
    axs[i].imshow(IMG)
plt.show()
    
```
